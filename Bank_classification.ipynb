```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import classification_report,confusion_matrix
import warnings
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv("Bank.csv")
data
```

```python
df = pd.DataFrame(data)
```

```python
df.isnull().sum()
```

```python
df.drop(['ID','CCAvg'] , axis= 1)
```

```python
X = pd.DataFrame(df , columns=['Age','Experience','Income','ZIP Code','Family','Education','Mortgage','Securities Account','CD Account','Online','CreditCard'])
Y = df['Personal Loan']
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y , test_size=0.15 ,random_state= 0)
```

```python
log = LogisticRegression( solver ='liblinear')
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
kfold_validation = KFold(10)
result =cross_val_score( log , X,Y , cv = kfold_validation )
print(result)
print(np.mean(result))
```

```python
data2 = df[df.index > 1500 ]
data3 = df[df.index < 2000 ]
data4 = pd.concat([data2 , data3])
```

```python
log.fit(X_train ,Y_train)
y_pred = log.predict(X_test)
```

```python
metrics.accuracy_score(Y_test , y_pred)
```

```python
print(confusion_matrix(Y , log.predict(X)))
```

```python
print(classification_report(Y , log.predict(X)))
```

```python
X_train2 , X_test2 , Y_train2 , Y_test2= train_test_split(X,Y , test_size=0.15 ,random_state= 0)
```

```python
K= 8
clf = KNeighborsClassifier()
clf.fit(X_train2 , Y_train2)
y_pred = clf.predict(X_test2)
print(metrics.accuracy_score(Y_test , y_pred))
```

```python
K = 50
acc = np.zeros((K))
for i in range (1,K+1):
    clf1 = KNeighborsClassifier(n_neighbors=i)
    clf1.fit(X_train2 ,Y_train2)
    y_pred2 =clf1.predict(X_test)
    acc[i-1]=metrics.accuracy_score(Y_test2 , y_pred2)
acc
```

```python
print(np.max(acc))
print(np.min(acc))
```
